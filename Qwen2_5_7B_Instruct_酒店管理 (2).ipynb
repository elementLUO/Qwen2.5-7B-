```python
import json
import os

# 获取当前工作目录（魔搭通常是 /mnt/workspace）
current_dir = os.getcwd()

# 1. 动态拼接文件路径（完美避开 /content 报错）
input_file = os.path.join(current_dir, "industry_instruction_semantic_cluster_dedup_住宿_餐饮_酒店_valid_val.jsonl")
output_path = os.path.join(current_dir, "LLaMA-Factory", "data")
output_file = os.path.join(output_path, "hotel_data.json")

# 确保目标文件夹存在
if not os.path.exists(output_path):
    os.makedirs(output_path)

converted_data = []

# 2. 尝试打开并转换
try:
    with open(input_file, 'r', encoding='utf-8') as f:
        for line in f:
            data = json.loads(line)
            convs = data.get("conversations", [])
            if len(convs) >= 2:
                converted_data.append({
                    "instruction": convs[0]["value"],
                    "input": "",
                    "output": convs[1]["value"]
                })

    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(converted_data, f, ensure_ascii=False, indent=2)
    print(f"✅ 成功转换数据: {len(converted_data)} 条，已存入 {output_file}")

    # 3. 自动注册到 dataset_info.json
    info_file = os.path.join(output_path, "dataset_info.json")
    
    # 魔搭环境下，先确认这个文件存不存在
    if os.path.exists(info_file):
        with open(info_file, "r", encoding="utf-8") as f:
            old_info = json.load(f)
    else:
        old_info = {} # 如果文件不存在，自动建一个空目录

    # 添加我们的酒店数据集
    old_info["hotel_zh_data"] = {"file_name": "hotel_data.json"}

    with open(info_file, "w", encoding="utf-8") as f:
        json.dump(old_info, f, ensure_ascii=False, indent=2)
    print("✅ 数据集已成功注册为 'hotel_zh_data'！")
    print("👉 下一步：你可以直接启动微调训练了！")

except FileNotFoundError:
    print(f"❌ 找不到原始 jsonl 文件！请确认你是否已经把它上传到了当前目录：{current_dir}")
```
